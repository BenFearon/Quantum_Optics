# Standard Quantum Limit in LIGO

LIGO scientists aim to decrease the noise floor of their interferometer by exploiting correlations between radiation pressure noise and shot noise.  These two components of noise are normally treated as uncorrelated, which gives a noise floor called the Standard Quantum Limit (SQL). By using squeezed light, and frequency dependent homodyne detection, third generation LIGO interferometers were able to beat the SQL.  Before delving into the mathematical underpinnings of the SQL, I'll briefly discuss LIGO's experimental setup and purpose.  


## LIGO 

LIGO is a gravitational wave observatory that detects gravitational waves through the use of a large, and highly callibrated interferometer. Cosmic events involving massive objects, like the combination of two black holes, can produce gravitational waves that propagate through space.  These waves distort the length of the interferometer's arms, leading to a detectable signal in the laser output.  LIGO's second observation run, between 11/2016 and 8/2017, saw the first detection of gravitational waves from black hole mergers and a neutron star inspiral.  Waveform reconstruction data for a binary black hole merger is shown below [2]:

## The SQL

LIGO's interferometric gravitational wave detectors experienced two primary sources of noise, photon shot noise, and radiation pressure noise.  

Photon shot noise is caused by the particle nature of light, which causes laser output data to be discrete as opposed to continuous.  Naturally, a higher rate of photon output would reduce shot noise, and thus, photon shot noise scales as the inverse of the laser's intensity.   

\begin{equation}
S_{shot} \propto \frac{1}{I_0}
\end{equation}

Radiation pressure noise is caused by photons applying unpredictable forces to the interferometer's mirror-endowed test masses. These forces cause the test masses to move in a manner that is unrelated to a gravitational wave event.  Radiation pressure noise thus scales proportionally to laser intensity.  

\begin{equation}
S_{rad} \propto I_0
\end{equation}

Without considering techniques using squeezed light, the minimum noise is achieved by a laser with power $I_{SQL}$ such that $S_{rad}=S_{shot}$. Thus, the noise floor is given by:

\begin{equation}
S \propto S_{rad} + S_{shot} = 2 S_{shot} \equiv h^2_{SQL}
\end{equation}

This minimum noise ($h^2_{SQL}$) is defined as the standard quantum limit. 

## Conventional Interferometer

### Input/Output Description
The image below depicts an interferometer with laser carrier $I_0$ entering through the bright port, and quantum fields a and b, respectively entering and leaving the dark port.  

<img src="files/if_basic.png" style="width: 250px;">

The general dark port input field is given by the expression:

\begin{equation}
E^+_{in} = \int_0^\infty \sqrt{\frac{2\pi\hbar\omega}{A c}}a_\omega e^{-i\omega t}\frac{d\omega}{dt}
\end{equation}

The annhilation operator $a_w$ obeys commutation relations $[a_\omega, a_\omega] = 0$, $[a_\omega, a_\omega^{\dagger}]=2\pi\delta(\omega-\omega ')$.

Splitting the field into sidebands about the carrier frequency $\omega_0$ gives $\omega = \omega_0 \pm \Omega.$  This treatment results in new creation and annhilation operators 

\begin{equation}
a_{\pm} \equiv a_{\omega \pm \Omega}
\end{equation}

\begin{equation}
[a_\pm, a_\pm] = 0, \ \ [a_\pm, a^{\dagger}_\pm] = 2\pi\delta(\Omega- \Omega ')
\end{equation}

Due to ponderomotive squeezing in the interferometer cavities, it is more convenient to represent the creation and annhilation operators in terms of the two-photon modes

\begin{equation}
a_{1} \equiv \frac{a_+ + a_-^{\dagger}}{\sqrt{2}}, \ \ \ \ a_{2} \equiv \frac{a_+ - a_-^{\dagger}}{\sqrt{2}i}
\end{equation}

\begin{equation}
[a_1, a^{\dagger}_2] = 2\pi i \delta(\Omega- \Omega ')
\end{equation}

Writing the dark port input in terms of $a_1$ and $a_2$ give the full electric field operator

\begin{equation}
E^+_{in} = \sqrt{\frac{4\pi\hbar\omega}{A c}}\cos(\omega_0 t)\int_0^\infty (a_1 e^{-i\Omega t} + a_1^{\dagger} e^{i\Omega t})\frac{d\Omega}{2\pi}
\end{equation}

\begin{equation}
E^-_{in} = \sqrt{\frac{4\pi\hbar\omega}{A c}}\sin(\omega_0 t)\int_0^\infty (a_2 e^{-i\Omega t} + a_2^{\dagger} e^{i\Omega t})\frac{d\Omega}{2\pi}
\end{equation}

\begin{equation}
E_{in} = E^+_{in} + E^-_{in}
\end{equation}

The dark port field operator will have the same form as the input field operator, but will use a parallel set of creation and annhilation operators denoted $b_1$ and $b_2$. By idealizing the beamsplitter as infinitessimally thin, and the arm cavities as integer multiples of the carrier wave-length, we find that and $b_1$ and $b_2$ can be related to $a_1$ and $a_2$ by:

\begin{equation}
b_1 = \Delta b_1 = a_1e^{2i\beta},\ \ b_2 = \Delta b_2 + \sqrt{2\kappa}\frac{h}{h_{SQL}}e^{i\beta},\ \ \Delta b_2 = (a_2-\kappa a_1)e^{2i\beta}
\end{equation}

Where $\Delta b_j$ components represent the noise components of $b_j$, and $\kappa$ is the rate at which $a_1$ and $b_2$ are coupled by radiation pressure back-action.  



### Gravitational Wave Effects:

To quantify the effect of gravitational waves and radiation pressure on the arm lengths ($X_n$ and $X_e$) of the interferometer, it's helpful to define a quantum observable x:

\begin{equation}
x \equiv X_n - X_e
\end{equation}

x evolves according to the expression below, where $h(t)$ is the contribution from the gravitational wave, and $x(t)_{BA}$ is the contribution from radiation pressure back-action. 

\begin{equation}
x(t)  = \int_{-\infty}^{\infty} e^{-i\Omega t} (L h(t) +x(t)_{BA})\frac{d\Omega}{2\pi}
\end{equation}

### Conventional Interferometer SQL

A conventional interferometer uses the vacuum state as the input to the dark port.  In this situation, the operators $a_1$ and $a_2$ are of the same form as fourier transforms of classical random processes

\begin{equation}
S_{a1}(f) = S_{a2}(f) = 1, \ \ \ S_{a1, a2}(f) = 0
\end{equation}

From equations (9) and (1), we can express the noise in the gravitational wave measurement as

\begin{equation}
h_n(\Omega) = \frac{h_{SQL}}{\sqrt{2\kappa}}\Delta b_2
\end{equation}

Finally, using equations (9), (12) and (15), we can obtain the noise spectral density for a conventional interferometer:

\begin{equation}
S_h = \frac{h^2_{SQL}}{2}(\frac{1}{\kappa} + \kappa)
\end{equation}

Tuning the laser power to $I_{SQL}$, minimizes $S_h$ at the interferometer's ideal operating frequency $(\Omega = \gamma)$, giving:

\begin{equation}
S_h = \frac{4\hbar}{mL^2\Omega^2}\left(\frac{2\gamma^4}{\Omega^2(\gamma^2+\Omega^2)}+\frac{\Omega^2(\gamma^2+\Omega^2)}{2\gamma^4}\right)
\end{equation}

The plot below shows $S_h$, normalized by $h_{SQL}$.  The curve's minimum is at the coordinate (1,1), intuitively signifying the laser's optimal performance at frequency $\Omega = \gamma$. 

```python
x = np.arange(.1, 10, .01)
Shc = 2/np.sqrt(8)*np.sqrt(2./(x**2+x**4)+(x**2+x**4)/2)
plt.plot(x, Shc, '-', label="Conventional IF")
plt.legend(framealpha=.5)
plt.yscale('log')
plt.xscale('log')
plt.xlabel('$\Omega/\gamma$')
plt.ylabel('$\sqrt{S_h(\Omega/\gamma)}/h_{SQL}$')
plt.title('IF Noise Floor vs. Frequency ($\Omega/\gamma$)')
axes = plt.gca()
axes.set_xlim([.01,10])
axes.set_ylim([.01,10])
plt.show()
```

## Ponderomotive Squeezing

Ponderomotive squeezing is an squeezing effect produced by the radiation pressure back-action on the mirrored test masses. Light traveling through the interferometer is thus squeezed, resulting in the input-output relation:

\begin{equation}
b_j = S^{\dagger}(r,\phi)R^{\dagger}(-\theta)a_je^{2i\beta}R(-\theta)S(r,\phi)
\end{equation}

Input light is rotated in the $a_1$ $a_2$ plane by angle $\theta$, and squeezed at angle $\phi$ by squeezing factor $r$.  Squeeze angle, rotation angle, and squeeze factor are related to $\kappa$, and by extension $\Omega$, as:

\begin{equation}
\theta = \arctan(\kappa/2),\ \ \phi = \frac{1}{2}arccot(\kappa/2),\ \ r = arcsinh(\kappa/2)
\end{equation}

The figure below depicts input vacuum at the dark port of the interferometer (left), and vacuum that exits the dark port, accompanying the interferometer signal (right).  


<img src="files/p_squeeze.png" style="width: 400px;">

The frequency dependence of the squeezing parameters $r$, $\theta$, and $\phi$, is an essential part of quantum non-demolition methods that beat $h_{SQL}$.

## Squeezed  Input Interferometer

The first QND approach discussed by Kimble et al. is the squeezed input interferometer. In this variation, a squeezed input is fed into the dark port of the interferometer.  

Thus, while a conventional interferometer's input state is given by vacuum state $|0_a\rangle$, the squeezed input interferometer's input state is given by 

\begin{equation}
|in\rangle = S(R,\lambda)|0_a\rangle
\end{equation}

where $R$ and $\lambda$ are respectively the maximum experimental squeeze factor, and an adjusted squeeze angle that minimizes the noise in the output quadrature $b_2$.  

Gravitational wave noise for the squeezed input interferometer is given by:

\begin{equation}
 h_n = \langle 0_a|h_{ns}h_{ns'}|0_a\rangle,\ \ \ h_{ns} = S^\dagger(R,\lambda)h_nS(R,\lambda)
\end{equation}

Equation (2) can be evaluated using the previously discussed expression for $h_n$ (equation 13), and the squeeze relation for $\Delta b_2$ (equation 9), giving:

\begin{equation}
 h_n = -\frac{h_{SQL}}{\sqrt{2\kappa}}\sqrt{1+\kappa^2}e^{i\beta} \times (a_1(\cosh(R)\cos(\Phi)-\sinh(R)\cos(\Phi-2(\Phi+\lambda)))-a_2(\cosh(R)\sin(\Phi)-\sinh(R)\sin(\Phi-2(\Phi+\lambda))))
\end{equation}

\begin{equation}
\Phi = arccot(\kappa)
\end{equation}

As in the case of the conventional interferometer, we treat $a_1$ and $a_2$ as classical semi-random processes, with spectral densities $S_{a_i} = 1$ and $S_{a_i,a_j} = 0$.  This treatment gives the spectral density of gravitational wave noise $S_h$:

\begin{equation}
S_h = \frac{h_{SQL}^2}{2}\left(\frac{1}{\kappa}+\kappa\right)(\cosh(2R) - \cos(2(\lambda + \Phi))\sinh(2R))
\end{equation}

This expression for $S_h$ is dependent on the input squeeze angle \lambda, which can be chosen to minimize $S_h$. Thus minimized, $S_h$ is given by:

\begin{equation}
S_h = \frac{h_{SQL}^2}{2}\left(\frac{1}{\kappa}+\kappa\right)e^{-2R} 
\end{equation}

Comparing this result with the noise spectral density of a conventional interferometer, it's clear that the squeezed input technique reduces the noise floor by a factor of e^{-2R}.  So, with an input laser power of $I_0 = I_{SQL}$, $S_h$ can be written as

\begin{equation}
S_h = \frac{4\hbar}{mL^2\Omega^2}\left(\frac{2\gamma^4}{\Omega^2(\gamma^2+\Omega^2)}+\frac{\Omega^2(\gamma^2+\Omega^2)}{2\gamma^4}\right)e^{-2R}
\end{equation}

where $e^{-2R} \approx 0.1$ experimentally.  The result is plotted below with the noise floor for a conventional interferometer for comparison.

```python
x = np.arange(.001, 10, .01)
Shc = 2/np.sqrt(8)*np.sqrt(2./(x**2+x**4)+(x**2+x**4)/2)
Shsi = 2/np.sqrt(8)*np.sqrt(2./(x**2+x**4)+(x**2+x**4)/2)*.1
plt.plot(x, Shc, '-', label="Conventional")
plt.plot(x, Shsi, '-', label="Squeezed Input")
plt.legend(framealpha=.5)
plt.yscale('log')
plt.xscale('log')
plt.xlabel('$\Omega/\gamma$')
plt.ylabel('$\sqrt{S_h(\Omega/\gamma)}/h_{SQL}$')
plt.title('IF Noise Floor vs. Frequency ($\Omega/\gamma$)')
axes = plt.gca()
axes.set_xlim([.01,10])
axes.set_ylim([.01,10])
plt.show()
```

Kimble et al. go on to describe a method for achieving the required frequency dependent squeeze angle
\begin{equation}
\lambda(\Omega) \equiv -\Phi(\Omega) = -arccot(\frac{2\gamma^4}{\Omega^2(\gamma^2+\Omega^2)})
\end{equation}
using a non-linear crystal and two Fabry-Prot cavities.  

## Variational Output Interferometer

The variational output interferometer supplies ordinary vacuum $|0_a\rangle$ to the dark port, but uses frequency dependent homodyne detection on the output field.  This is to say that instead of measuring $b_2$, the variational output interferometer measures:
\begin{equation}
b_\xi = b_1\cos(\xi)+b_2\sin(\xi)
\end{equation}

where $\xi$ is dependent on $\Omega$.  

Using the expressions:
\begin{equation}
b_1 = \Delta b_1 = a_1e^{2i\beta},\ \ b_2 = \Delta b_2 + \sqrt{2\kappa}\frac{h}{h_{SQL}}e^{i\beta},\ \ \Delta b_2 = (a_2-\kappa a_1)e^{2i\beta}
\end{equation}

we can find $b_\xi$ in terms of the interferometer's dark port inputs and the gravitational wave field.

\begin{equation}
b_\xi = \sin(\xi)\left(\sqrt{2\kappa}\frac{h}{h_{SQL}}e^{i\beta}+(a_2+a_1(\cot(\xi)-\kappa))e^{2i\beta}\right)
\end{equation}

Using the definition of $h$ gives the operator for the fourier transform of the system's gravitational wave noise:

\begin{equation}
 h_n = \frac{h_{SQL}}{\sqrt{2\kappa}}e^{i\beta}(a_2+a_1(cot(\xi_-\kappa))
\end{equation}

Clearly, choosing $\xi=arccot(\kappa)$ eliminates all back-action induced noise from the system output. Minimizing noise in this manner gives the expression

\begin{equation}
S_h = \frac{4\hbar}{mL^2\Omega^2}\frac{1}{I_0/I_{SQL}}\left(\frac{\Omega^2(\gamma^2+\Omega^2)}{2\gamma^4}\right)
\end{equation}

for spectral noise density for a variational output interferometer.  This noise floor is plotted below, along with the noise floors for squeezed input and conventional interferometers for comparison. 

```python
x = np.arange(.001, 10, .01)
Shc = 2/np.sqrt(8)*np.sqrt(2./(x**2+x**4)+(x**2+x**4)/2)
Shsi = 2/np.sqrt(8)*np.sqrt(2./(x**2+x**4)+(x**2+x**4)/2)*.1
Shvo = 2/np.sqrt(8)*np.sqrt((x**2+x**4)/2)*.1

plt.plot(x, Shc, '-', label="Conventional")
plt.plot(x, Shsi, '-', label="Squeezed Input")
plt.plot(x, Shvo, '-', label="Variational Output")

plt.legend(framealpha=.5)
plt.yscale('log')
plt.xscale('log')
plt.xlabel('$\Omega/\gamma$')
plt.ylabel('$\sqrt{S_h(\Omega/\gamma)}/h_{SQL}$')
plt.title('IF Noise Floor vs. Frequency ($\Omega/\gamma$)')
axes = plt.gca()
axes.set_xlim([.01,10])
axes.set_ylim([.001,10])
plt.show()
```

## Squeezed Variational Inteferometer

As its name suggests, the squeezed variational interferometer combines the defining aspects of the squeezed-input interferometer and the variational output interferometer to beat the SQL to a greater extent than either of them.  

Since a squeezed variational interferometer measures a frequency dependent output quadrature $b_\xi$, the gravitational wave noise operator $h_n$ is the same as in the variational output interferometer case:

\begin{equation}
 h_n = \frac{h_{SQL}}{\sqrt{2\kappa}}e^{i\beta}(a_2+a_1(cot(\xi_-\kappa))
\end{equation}

Plugging $h_n$ into $\langle 0_a|S^\dagger h_n S|0_a\rangle$ and treating $a_1$ and $a_2$ as classical random processes gives:

\begin{equation}
S_h = \frac{h_{SQL}^2}{2\kappa}\left(1+\kappa'^2\right)(e^{-2R}+\sinh(2R)(1- \cos(2(\lambda + \Phi '))
\end{equation}

where $\kappa ' = \kappa - \cot(\xi)$ and $\Phi ' = arccot(\kappa ')$.

$S_h$ is minimized by setting $\lambda = \pi/2$ and $\xi = \Phi = arccot(\kappa)$ which gives:

\begin{equation}
S_h = \frac{4\hbar}{mL^2\Omega^2}\frac{e^{-2R}}{I_0/I_{SQL}}\left(\frac{\Omega^2(\gamma^2+\Omega^2)}{2\gamma^4}\right)
\end{equation}

It's evident that this noise floor incorporates the factors $e^{-2R}$ and $I_0/I_{SQL}$ that are the result of squeezing the input, and performing heterodyne detection at the output, respectively. Again, the results are plotted below for comparison.  


```python
x = np.arange(.001, 10, .01)
Shc = 2/np.sqrt(8)*np.sqrt(2./(x**2+x**4)+(x**2+x**4)/2)
Shsi = 2/np.sqrt(8)*np.sqrt(2./(x**2+x**4)+(x**2+x**4)/2)*.1
Shvo = 2/np.sqrt(8)*np.sqrt((x**2+x**4)/2)*.1
Shsv = 2/np.sqrt(8)*np.sqrt((x**2+x**4)/2)*.1*.1

plt.plot(x, Shc, '-', label="Conventional")
plt.plot(x, Shsi, '-', label="Squeezed Input")
plt.plot(x, Shvo, '-', label="Variational Output")
plt.plot(x, Shsv, '-', label="Squeezed Variational")


plt.legend(framealpha=.5)
plt.yscale('log')
plt.xscale('log')
plt.xlabel('$\Omega/\gamma$')
plt.ylabel('$\sqrt{S_h(\Omega/\gamma)}/h_{SQL}$')
plt.title('IF Noise Floor vs. Frequency ($\Omega/\gamma$)')
axes = plt.gca()
axes.set_xlim([.01,10])
axes.set_ylim([.001,10])
plt.show()
```
